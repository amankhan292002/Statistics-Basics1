1.  What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
    - Statistics is broadly divided into descriptive statistics and inferential statistics, based on how data is used and interpreted.

    Descriptive Statistics

    Descriptive statistics deals with summarizing and describing data that is already collected. It does not make predictions or generalizations beyond the given data.
    It uses numerical measures and visual tools to present data clearly.

    Common techniques:

    Mean, median, mode

    Range, variance, standard deviation

    Tables, charts, graphs

    Example:
    If we calculate the average marks of students in a class, we are using descriptive statistics.

    - Inferential statistics uses a sample of data to make conclusions or predictions about a larger population. It involves uncertainty and probability.

    Common techniques:

    Hypothesis testing

    Confidence intervals

    Regression analysis

    Example:
    If we use marks of 50 students to predict the average marks of all students in a university, we are using inferential statistics.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

   - Sampling in Statistics

    Sampling is the process of selecting a subset (sample) from a large population to analyze and draw conclusions.
    It saves time, cost, and effort while still providing reliable results.

    -> Random Sampling

    In random sampling, every individual has an equal chance of being selected.
    Selection is completely unbiased.

    Characteristics:

    Simple to implement

    No guarantee of subgroup representation

    Suitable for homogeneous populations

    Example:
    Selecting 50 students randomly from a college using a lottery method.

   -> Stratified Sampling

    In stratified sampling, the population is divided into subgroups (strata) based on characteristics such as age, gender, or income.
    Samples are then taken from each subgroup.

    Characteristics:

    Ensures representation of all groups

    More accurate than random sampling

    Slightly complex to implement

    Example:
    Dividing students into science and arts streams and selecting samples from both.

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

   - Measures of central tendency are statistical values that represent the center or typical value of a dataset. The three main measures are mean, median, and mode.

    -> Mean

    The mean is the average of all values in a dataset. It is calculated by dividing the sum of all observations by the total number of observations.

    Example:
    If marks are 60, 70, and 80, the mean is (60 + 70 + 80) / 3 = 70.

    -> Median

    The median is the middle value when the data is arranged in ascending or descending order. If there is an even number of observations, the median is the average of the two middle values.

    Example:
    For  values 10, 20, 30 → Median = 20

    -> Mode

    The mode is the value that occurs most frequently in the dataset. A dataset may have one mode, multiple modes, or no mode.

    Example:
    In 2, 4, 4, 6 → Mode = 4

    -> Importance of Measures of Central Tendency

    They summarize large datasets into a single representative value.

    They help in comparing different datasets.

    Mean is useful for mathematical analysis, median is useful when data has outliers, and mode is useful for categorical or repetitive data.

4.  Explain skewness and kurtosis. What does a positive skew imply about the
data?

    - Skewness

     Skewness measures the asymmetry of a data distribution.

    Zero skewness: Symmetrical distribution

    -> Positive skewness: Tail is longer on the right side

    -> Negative skewness: Tail is longer on the left side

    -> Positive Skewness

    A positive skew indicates that:

    Most data values are concentrated on the left

    A few large values (outliers) pull the tail to the right

    Mean > Median > Mode

    Example:
    Income distribution is often positively skewed because a few people earn very high incomes.

    Kurtosis

    -> Kurtosis measures the peakedness or flatness of a distribution.

    High kurtosis: Sharp peak and heavy tails (more outliers)

    Low kurtosis: Flat peak and light tails

```python
# 5. Implement a Python program to compute the mean, median, and mode of
# a given list of numbers.
# numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
# (Include your Python code and output in the code box below.)

#      - numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
#6.  Compute the covariance and correlation coefficient between the following
# two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]
# (Include your Python code and output in the code box below.)


# list_x = [10, 20, 30, 40, 50]
# # list_y = [15, 25, 35, 45, 60]
# Covariance measures how two variables change together.

# Positive covariance → variables increase together

# Negative covariance → one increases while the other decreases

# Correlation coefficient measures the strength and direction of the relationship and lies between –1 and +1.


import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance
covariance = np.cov(list_x, list_y)

# Correlation coefficient
correlation = np.corrcoef(list_x, list_y)

print("Covariance :\n", covariance)
print("\nCorrelation Coefficient :\n", correlation)
```

```python
# 7: Write a Python script to draw a boxplot for the following numeric list and
# identify its outliers. Explain the result:
# data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
# (Include your Python code and output in the code box below.)

import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()


# Result Explanation

# Most data values lie between 14 and 26.

# The value 35 appears as an outlier because it lies far above the upper whisker.

# This indicates the dataset is right-skewed, with one unusually high value.
```

```python
#  8: You are working as a data analyst in an e-commerce company. The
# marketing team wants to know if there is a relationship between advertising spend and
# daily sales.
# ● Explain how you would use covariance and correlation to explore this
# relationship.
# ● Write Python code to compute the correlation between the two lists:
# advertising_spend = [200, 250, 300, 400, 500]
# daily_sales = [2200, 2450, 2750, 3200, 4000]
# (Include your Python code and output in the code box below.)


import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Covariance
covariance = np.cov(advertising_spend, daily_sales)

# Correlation
correlation = np.corrcoef(advertising_spend, daily_sales)

print("Covariance Matrix:\n", covariance)
print("\nCorrelation Coefficient Matrix:\n", correlation)



#Explanation:
# The positive covariance shows that advertising spend and sales increase together.

# The correlation coefficient (~0.99) indicates a very strong positive relationship.

# This suggests that higher advertising spend leads to higher daily sales, supporting data-driven marketing decisions.
```

```python
#  9: Your team has collected customer satisfaction survey data on a scale of
# 1-10 and wants to understand its distribution before launching a new product.
# ● Explain which summary statistics and visualizations (e.g. mean, standard
# deviation, histogram) you’d use.
# ● Write Python code to create a histogram using Matplotlib for the survey data:
# survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
# (Include your Python code and output in the code box below.)





import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

print("Mean:", np.mean(survey_scores))
print("Standard Deviation:", np.std(survey_scores))

plt.hist(survey_scores, bins=6, edgecolor='black')
plt.xlabel("Survey Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Satisfaction Scores")
plt.show()
```
